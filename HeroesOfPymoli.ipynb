```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(10)
```

## Player Count

* Display the total number of players


```python
new_df = purchase_data[["SN"]]
new_df = new_df.rename(columns={"SN": "Total Players"})
new_df.nunique()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
d = {"Number of Unique Items": [purchase_data["Item ID"].nunique()], 
     "Average Price":[round(purchase_data["Price"].mean(), 2)],
    "Number of Purchases": [purchase_data.shape[0]],
    "Total Revenue" : purchase_data["Price"].sum()}
pd.options.display.float_format = '${:,.2f}'.format
P_Analysis = pd.DataFrame(d)
P_Analysis 
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
try_df = purchase_data[["SN", "Age", "Gender"]].copy()
try_df.drop_duplicates(subset="SN", keep='first', inplace=True)
count_males = len(try_df.loc[(try_df["Gender"] == "Male")])
count_females = len(try_df.loc[(try_df["Gender"] == "Female")])
count_other = len(try_df.loc[(try_df["Gender"] == "Other / Non-Disclosed")])
total_all_genders = len(try_df)
d = {"Total Count" : [count_males, count_females, count_other],
    "Percentage of Players" : [(count_males/total_all_genders) * 100, 
                                (count_females/total_all_genders) * 100,
                                (count_other/total_all_genders) * 100]}
pd.options.display.float_format = '{:,.2f}'.format
new_df = pd.DataFrame(d, index=["Male", "Female", "Other / Non-Disclosed"])
new_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Extract the following columns: "Gender", "Price"
p_type = purchase_data[["Gender", "Price"]].copy()
p_type.rename(columns={"Price": "Average Purchase Price"}, inplace=True)
p_group = p_type.groupby(["Gender"])
p_group_avg = p_group.mean()
p_group_avg["Total Purchase Value"] = p_group.sum()
p_group_avg["Purchase Count"] = purchase_data["Gender"].value_counts()
# 
# Using calculations from above, create a list of avg total purchase per unique person
#
p = [round((p_group_avg.iloc[0,1]/count_females), 2), round((p_group_avg.iloc[1,1]/count_males), 2), 
     round((p_group_avg.iloc[2,1]/count_other), 2)]

p_group_avg["Avg Total Purchase per Person"] = p
p_group_avg = p_group_avg[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]
pd.options.display.float_format = '${:,.2f}'.format
p_group_avg
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
df_ages = pd.DataFrame(try_df["Age"])

categories = pd.cut(df_ages["Age"], [0, 9, 14, 19, 24, 29, 34, 39, 100], labels=['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+'])
gdf = df_ages.groupby(categories)["Age"].agg(['count'])
gdf.rename(columns={"count": "Total Count"}, inplace=True)
gdf["Percentage of Players"] = round((gdf["Total Count"]/len(try_df)) * 100, 2)
pd.options.display.float_format = '{:,.2f}'.format
gdf
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
price_ages = purchase_data[["Age", "Price"]].copy()
categories = pd.cut(price_ages["Age"], [0, 9, 14, 19, 24, 29, 34, 39, 100], labels=['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+'])
pricedf = price_ages.groupby(categories)["Price"].agg(['mean'])
pricedf.rename(columns={"mean": "Average Purchase Price"}, inplace=True)
pricedf["Purchase Count"] = price_ages.groupby(categories)["Age"].agg(['count'])
pricedf["Total Purchase Value"] = price_ages.groupby(categories)["Price"].agg(['sum'])
pricedf["Avg Total Purchase Per Person"] = pricedf["Total Purchase Value"]/gdf["Total Count"]
pricedf = pricedf[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase Per Person"]]
pd.options.display.float_format = '${:,.2f}'.format
pricedf
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
ts = purchase_data[["SN", "Price"]].copy()
t2 = purchase_data.groupby(ts["SN"])["Price"].agg(['sum'])
t2["Purchase Count"] = purchase_data.groupby(ts["SN"])["Price"].agg('size')
t2["Average Purchase Price"] = purchase_data.groupby(ts["SN"])["Price"].agg('mean')
t2.rename(columns={"sum": "Total Purchase Value"}, inplace=True)
t2 = t2[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
t2 = t2.sort_values(by="Total Purchase Value", ascending=False)
t2.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_df = purchase_data[["Item ID", "Item Name", "Price"]].copy()
# calculate Total Purchase Value

i2count = item_df.groupby(["Item ID", "Item Name"]).count()["Price"]
item_price = item_df.groupby(["Item ID", "Item Name"]).mean()["Price"]
i2sum = item_df.groupby(["Item ID", "Item Name"]).sum()["Price"]

together = pd.DataFrame({"Purchase Count":i2count, "Item Price": item_price, "Total Purchase Value":i2sum})
t1 = together.sort_values(by="Purchase Count", ascending=False)
t1.head()
```

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting
